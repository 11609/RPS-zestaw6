```python
import random

def U():
  return random.uniorm(0, 1)

def bernouli(p):
  return U() < p

def binom(n, p):
  return sum(bernouli(p) for i in range(n))
```

Kod i demonstracja:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

DYSTRYBUCJA = [0.25, 0.15, 0.1, 0.2, 0.3] # = [p(0), p(1), ..., p(4)]

# rnd losuje zmienną zgodną z DYSTRYBUCJĄ
def rnd():
  uni = np.random.uniform(0, sum(DYSTRYBUCJA))
  for i in range(len(DYSTRYBUCJA)):
    if uni <= DYSTRYBUCJA[i]:
      return i
    uni -= DYSTRYBUCJA[i]

# sum_rnd losuje sumę 'n' zmiennych,
# gdzie każda jest zgodna z DYSTRYBUCJĄ
def sum_rnd(n):
  return sum(rnd() for i in range(n))

# estim_sum_dist estymuję dystrybucję zmiennej
# sum_rnd(n) poprzez wielokrotne losowanie jej m razy
def estim_sum_dist(n, m):
  freq = len(DYSTRYBUCJA) * n * [0]
  for i in range(m):
    freq[sum_rnd(n)] += 1
  return freq

if __name__ == '__main__':
  n = int(input("Ile zmiennych zasymulować? "))
  m = int(input("Ile razy losować sumę zmiennych? "))

  values = estim_sum_dist(n, m)
  plt.plot(values)
  plt.show()
```

## Eksperyment dla n = 5:

```python
values = estim_sum_dist(5, 1000)
plt.plot(values)
plt.show()
```
